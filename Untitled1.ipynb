```python
import pandas as pd

def read_all_sheets_from_excel(path: str) -> dict:
    dataframes = pd.read_excel(path, sheet_name=None)
    return dataframes

# Read the Excel file
file_path = 'data.xlsx'
dataframes = read_all_sheets_from_excel(file_path)

# print each dataframe name
print('Dataframes in this file:', ', '.join(dataframes.keys()))

for k, v in dataframes.items():
    # strip whitespace where possible from column names; need to check if isinstance(x, str) because some column names are numbers
    try:
        v = v.rename(columns=lambda x: x.strip() if isinstance(x, str) else x)
    except:
        pass

    # strip whitespace where possible from cells
    try:
        v = v.apply(lambda col: col.str.strip() if col.dtype == 'object' else col)
    except:
        pass
    dataframes[k] = v
    print('dataframe: '+ k)
    print(v.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = 'data.xlsx'
df = pd.read_excel(file_path, sheet_name='Sheet1')

# Univariate Analysis
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Plot PDFs, Histograms, Boxplots for numerical columns
for col in numerical_cols:
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 3, 1)
    sns.histplot(df[col], kde=True, color='skyblue')
    plt.title(f'Histogram of {col}')
    
    plt.subplot(1, 3, 2)
    sns.boxplot(y=df[col], color='lightblue')
    plt.title(f'Boxplot of {col}')
    
    plt.subplot(1, 3, 3)
    sns.kdeplot(df[col], color='salmon', shade=True)
    plt.title(f'PDF of {col}')
    
    plt.tight_layout()
    plt.show()
```

```python
#B

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
# Assuming the dataset is already loaded in df from previous operations

# Bivariate Analysis

# 1. Relationships between numerical columns
# Scatter plot between 10th percentage and Salary
df_sample = df.sample(n=500, random_state=1) # Sample for quicker plotting
plt.figure(figsize=(10, 6))
sns.scatterplot(x='10percentage', y='Salary', data=df_sample)
plt.title('10th Percentage vs Salary')
plt.xlabel('10th Grade Percentage')
plt.ylabel('Salary')
plt.show()

# 2. Categorical and numerical columns
# Boxplot for Gender vs Salary
df_sample = df.sample(n=500, random_state=1) # Sample for quicker plotting
plt.figure(figsize=(10, 6))
sns.boxplot(x='Gender', y='Salary', data=df_sample)
plt.title('Gender vs Salary')
plt.xlabel('Gender')
plt.ylabel('Salary')
plt.show()

# 3. Categorical and categorical columns
# Stacked bar plot for Gender and JobRole
job_role_gender = df.groupby(['Gender', 'Designation'])['ID'].count().unstack().fillna(0)
job_role_gender.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Job Role Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
#Examining the relationship between job location and salary using visualizations

import matplotlib.pyplot as plt

# Grouping the data by JobCity and calculating the average salary
salary_by_city = df.groupby('JobCity')['Salary'].mean().sort_values(ascending=False)

# Plotting the relationship between JobCity and average Salary
plt.figure(figsize=(12, 6))
plt.bar(salary_by_city.index[:10], salary_by_city.values[:10])
plt.xlabel('Job City')
plt.ylabel('Average Salary')
plt.title('Average Salary by Job City')
plt.xticks(rotation=45)
plt.show()
```

```python
# Salaries claim case and gender specific job scenarios
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
file_path = 'data.xlsx'
df = pd.read_excel(file_path)

# Filter data for the specified roles and calculate average salary
roles = ['programming analyst', 'software engineer', 'hardware engineer', 'associate engineer']
df_filtered = df[df['Designation'].str.lower().isin(roles)]
avg_salary = df_filtered['Salary'].mean()

# Test the claim
claim_test_result = 'meets' if 250000 <= avg_salary <= 300000 else 'does not meet'

# Gender and Specialization relationship
# Creating a contingency table
contingency_table = pd.crosstab(df['Gender'], df['Specialization'])

# Chi-square test of independence
from scipy.stats import chi2_contingency
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Plotting
plt.figure(figsize=(12, 8))
sns.countplot(x='Specialization', hue='Gender', data=df)
plt.title('Gender vs Specialization')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Results
avg_salary, claim_test_result, p
```
